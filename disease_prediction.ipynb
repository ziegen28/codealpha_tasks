#Importing Libraries

```python
import numpy as np
import pandas as pd
```

#Reading the dataset

```python
train=pd.read_csv('/content/Training.csv')
train.head()
```

#Data preparation

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
train['prognosis']=encoder.fit_transform(train.prognosis)
```

```python
train=train.drop(['Unnamed: 133'],axis=1)
train.head()
```

```python
train.corr()
```

```python
train.isnull().sum()
```

#Random Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X = train.iloc[:,:-1]
y = train.iloc[:, -1]
```

```python
rfc=RandomForestClassifier()
rfc.fit(X,y)
```

```python
test=pd.read_csv('/content/Testing.csv')
test.head()
```

```python
test['prognosis']=encoder.transform(test.prognosis)
```

```python
testx=test.iloc[:,:-1]
testy=test.iloc[:,-1]

y_pred=rfc.predict(testx)
```

#model evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
print(f"Accuracy on train data by Random Forest Classifier\: {accuracy_score(y, rfc.predict(X))*100}")

print(f"Accuracy on test data by Random Forest Classifier\: {accuracy_score(testy, y_pred)*100}")
```

```python
cf_matrix = confusion_matrix(testy, y_pred)
plt.figure(figsize=(12,8))
sns.heatmap(cf_matrix, annot=True)
plt.title("Confusion Matrix for Random Forest Classifier on Test Data")
plt.show()
```
